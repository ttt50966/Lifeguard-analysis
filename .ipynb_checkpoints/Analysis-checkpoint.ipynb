```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import IntProgress
from IPython.display import display
import matplotlib as mpl
import locale
mpl.rc("font", family = "Noto Sans CJK TC")
import time
```

```python
data = pd.read_csv('data.csv')
data[(data["證照號碼"].str[:2] == '水協') & (data["證照生效日"].str[:4] == '2018')][["證照資格","證照號碼","證照生效日"]]
```

```python
years = ['2016','2017','2018']
group_names = ["水協","海爆", "紅會", "水運", "水域", "海龍","海總","游救"]
LifeSaving = []
Sea = []
Red = []
Sport =[]
Area = []
Dragon = []
Sea_A = []
Swim =[]
group = {"水協":LifeSaving, "海爆":Sea, "紅會":Red,
         "水運":Sport, "水域":Area, "海龍":Dragon, "海總":Sea_A,
         "游救":Swim}
```

```python
for year in years:
    for g in group_names:
        group[g].append(len(data[(data["證照號碼"].str[:2] == g) & (data["證照生效日"].str[:4] == year)]))
```

```python
group
```

```python
plt.figure(dpi=300)
for g in group_names:
    plt.plot(years,group[g], label=g)
plt.title("體育署救生員證人數")
plt.xlabel("年度")
plt.ylabel("人數 (人)")
plt.legend(loc=(1.01,0.43))
plt.tight_layout(pad=0.4, w_pad=1.2, h_pad=1.0)
#plt.savefig("Lifeguard_number.png")
```

```python
year_sum = [0]*3
for i in range(3):
    year_sum[i] = len(data[data["證照生效日"].str[:4] == years[i]])
year_sum
```

```python
plt.figure(dpi=300)
for g in group_names:
    plt.plot(years,np.array(group[g])/np.array(year_sum)*100, label=g)
plt.legend(loc=(1.01,0.43))
plt.title("體育署救生員證佔當年度百分比")
plt.xlabel("年度")
plt.ylabel("百分比 (%)")
plt.tight_layout(pad=0.4, w_pad=1.2, h_pad=1.0)
#plt.savefig("Lifeguard_percentage.png")
```
